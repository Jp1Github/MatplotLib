## Code

```python
'''Setting the background color of the plot using  set_facecolor method'''
ax = plt.axes()
ax.set_facecolor('#D3E4CD')


'''Label the title, x-axis and y-axis'''
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')


'''Plotting'''
plt.plot(ages_x, dev_y, color = '#ff800e', linewidth = 2)
plt.plot(ages_x, py_dev_y, color = '#008176', linewidth = 3)
# plt.plot(ages_x, js_dev_y, color = '#0000a7')
plt.legend(['All Developer', 'Python', 'JavaScript'], facecolor='#EFFFFD')


'''Filling using the fill_between Method.
It will fill from Python to the zero axis'''
# plt.fill_between(ages_x, py_dev_y)
# By default y is set zero(0)
# plt.fill_between(ages_x, py_dev_y)


# alpho argument set the opacity. Range 0-1 which 1 no opacity.
plt.fill_between(ages_x, py_dev_y, alpha = 0.5)

'''Putting padding'''
plt.tight_layout()

'''Putting a grid'''
plt.grid(color = '#A3C6C4',  linestyle = 'dotted', linewidth = 0.8)


'''To show the graph'''
plt.show()

```

### Fill set a specific number

```python
'''Setting the background color of the plot using  set_facecolor method'''
ax = plt.axes()
ax.set_facecolor('#D3E4CD')


'''Label the title, x-axis and y-axis'''
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')


'''Plotting'''
plt.plot(ages_x, dev_y, color = '#ff800e', linewidth = 2)
plt.plot(ages_x, py_dev_y, color = '#008176', linewidth = 3)
# plt.plot(ages_x, js_dev_y, color = '#0000a7')
plt.legend(['All Developer', 'Python', 'JavaScript'], facecolor='#EFFFFD')

'''Create a median value'''
Median_Salary = 57287


# alpho argument set the opacity. Range 0-1 which 1 no opacity.
plt.fill_between(ages_x, py_dev_y, Median_Salary, alpha = 0.2)

'''Putting padding'''
plt.tight_layout()

'''Putting a grid'''
plt.grid(color = '#A3C6C4',  linestyle = 'dotted', linewidth = 0.8)


'''To show the graph'''
plt.show()

```

### Color Format Area

```python
'''Setting the background color of the plot using  set_facecolor method'''
ax = plt.axes()
ax.set_facecolor('#D3E4CD')


'''Label the title, x-axis and y-axis'''
plt.xlabel('Age')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')


'''Plotting'''
plt.plot(ages_x, dev_y, color = '#ff800e', linewidth = 2)
plt.plot(ages_x, py_dev_y, color = '#008176', linewidth = 3)
# plt.plot(ages_x, js_dev_y, color = '#0000a7')
plt.legend(['All Developer', 'Python', 'JavaScript'], facecolor='#EFFFFD')

'''Create a median value'''
Median_Salary = 57287


'''Convert it to an np array.
Else this error will pop up '>' not supported between instances of 'list' and 'int'''
y2 = np.array(Median_Salary)
y1 = np.array(py_dev_y)

# alpho argument set the opacity. Range 0-1 which 1 no opacity.
plt.fill_between(ages_x, py_dev_y, Median_Salary, where=(y1 > y2), alpha = 0.2,
                 interpolate = True)

plt.fill_between(ages_x, py_dev_y, Median_Salary, where=(y1 <= y2), alpha = 0.2,
                 interpolate = True, color = '#c1272d') ##c1272d - Red

'''Putting padding'''
# plt.tight_layout()

'''Putting a grid'''
plt.grid(color = '#A3C6C4',  linestyle = 'dotted', linewidth = 0.8)


'''To show the graph'''
plt.show()

```
